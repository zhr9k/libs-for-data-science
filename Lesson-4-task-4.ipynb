Посчитайте с помощью метода value_counts количество наблюдений для каждого значения целевой переменной Class и примените к
полученным данным метод plot, чтобы построить столбчатую диаграмму. Затем постройте такую же диаграмму, используя логарифмический масштаб.

```python
values_for_class = df.Class.value_counts()
values_for_class
```

```python
plt.figure(figsize=(10, 3))

plt.subplot(121)
values_for_class.plot(kind="bar", color='grey')
plt.xlabel("Class")
plt.ylabel("Counts")

plt.subplot(122)
values_for_class.plot(kind="bar", logy=True, color='grey')
plt.xlabel('Class')
plt.ylabel("Log scale")

plt.show()
```

На следующем графике постройте две гистограммы по значениям признака V1 - одну для мошеннических транзакций (Class равен 1) и другую - для обычных (Class равен 0). Подберите значение аргумента density так, чтобы по вертикали графика было расположено не число наблюдений, а плотность распределения. Число бинов должно равняться 20 для обеих гистограмм, а коэффициент alpha сделайте равным 0.5, чтобы гистограммы были полупрозрачными и не загораживали друг друга.
Создайте легенду с двумя значениями: “Class 0” и “Class 1”. Гистограмма обычных транзакций должна быть серого цвета, а мошеннических - красного. Горизонтальной оси дайте название “Class”.

```python
v1_good = df[['Class', 'V1']].query('Class == 0').V1
v1_fraud = df[['Class', 'V1']].query('Class == 1').V1
```

```python
plt.hist(v1_good, density=True, bins=20, color='grey', alpha=0.5, label='Class 0')
plt.hist(v1_fraud, density=True, bins=20, color='red', alpha=0.5, label='Class 1')
plt.xlabel('V1 values for Class')

plt.legend(loc='upper left')
plt.show()
```
