### Задание 2
С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.<br>

```python
t = np.linspace(0, 10, 51)
t
```

Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.<br>

```python
f = np.cos(t)
f
```

Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив f - для координат по вертикали.<br> Линия графика должна быть зеленого цвета.

```python
plt.plot(t, f, color="green")
plt.show()
```

Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.<br>
Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.<br>

```python
plt.plot(t, f, color="Green")
plt.title("График f=cos(t)")
plt.xlabel("Переменная t")
plt.ylabel("Значения f(t)")
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

### Задание 3

С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.<br>

```python
x = np.linspace(-3, 3, 51)
```

<sp>Создайте массивы y1, y2, y3, y4 по следующим формулам:<br>
y1 = x**2<br>
y2 = 2 * x + 0.5<br>
y3 = -3 * x - 1.5<br>
y4 = sin(x)<br>

```python
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
```

Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.<br>Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке,<br> состоящей из двух строк и двух столбцов. В каждом графике массив x используется для координат по горизонтали.<br>В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.<br>Дайте название графикам: 'График y1', 'График y2' и т.д.<br>
Для графика в левом верхнем углу установите границы по оси x от -5 до 5.<br>
Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.<br>
Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.<br>
<br>
<i>P.S. я тут немного поменял названия для графиков на чуть, как мне кажется, более информативные</i>

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
graph1, graph2, graph3, graph4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

graph1.plot(x, y1)
graph1.set_xlim([-5, 5])
graph1.set_title("y=x^2")

graph2.plot(x, y2)
graph2.set_title("y=2x+0.5")

graph3.plot(x, y3)
graph3.set_title("y=-3x-1.5")

graph4.plot(x, y4)
graph4.set_title("y=sin(x)")
```

### Задание 9
Создать одномерный массив а, состоящий из последовательности целых чисел от 0 до 12. Поменять форму этого массива, чтобы получилась матрица A (двумерный массив Numpy), состоящая из 4 строк и 3 столбцов.

```python
A = np.arange(0, 12).reshape(4, 3)
A
```

Получить матрицу At путем транспонирования матрицы A.

```python
At = A.T
At
```

Получить матрицу B, умножив матрицу A на матрицу At с помощью матричного умножения.

```python
B = A.dot(At)
B
```

Какой размер имеет матрица B?

```python
B.shape
```

Получится ли вычислить обратную матрицу для матрицы B и почему?<br>
<br>
Квадратная матрица обратима тогда и только тогда, когда ее определитель не равен нулю. 

```python
d = np.linalg.det(B)
d
```

Определитель В равен нулю, то есть не получится вычислить матрицу, обратную матрице В
