```python
def reduce_mem_usage(df):
    """ iterate through all the columns of a dataframe and modify the data type
        to reduce memory usage.        
    """
    start_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage of dataframe is {:.2f} MB'.format(start_mem))
    
    for col in df.columns:
        col_type = df[col].dtype
        
        if col_type != object:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            else:
                if c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
        else:
            df[col] = df[col].astype('category')

    end_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage after optimization is: {:.2f} MB'.format(end_mem))
    print('Decreased by {:.1f}%'.format(100 * (start_mem - end_mem) / start_mem))
    
    return df
```

```python
df.loc[df['HouseFloor'] == 0, ['HouseFloor', 'Floor']]
```

Сохраню индексы, чтоб проверить потом, что все норм получилось

```python
indexes_for_check = df.loc[df['HouseFloor'] == 0, ['HouseFloor', 'Floor']].index[:10]
```

Меня немного смущает, что HouseFloor имеет тип действительного числа, поэтому приведу его к инту

```python
df['HouseFloor'] = df['HouseFloor'].astype(np.int16)
```

```python
df.loc[indexes_for_check, ['HouseFloor', 'Floor']]
```

Будем считать, что с нулевым этажом разобрались. Теперь разберемся с ситуациями, когда текущий этаж больше общего количества этажей. И сразу встают вопросы: что брать за независимую переменную и к какому значению менять?<br>
За независимую переменную возьму общее количество этажей, потому что мне кажется, что вероятность ошибки у значения количества этажей в самом доме меньше, чем вероятность ошибки у значения этажа квартиры<br>
Теперь надо понять, какие значения ставить тем строкам, у которых $Floor > HouseFloor$. Самый простой вариант - сделать Floor таким же, как и HouseFloor, как это было сделано для случаев, когда $HouseFloor = 0$. Но у меня есть предположение, что "высота" квартиры (отношение этажа квартиры к общему числу этажей), поэтому сделаю новую фичу $FloorPos = \frac{Floor}{HouseFloor}$

```python
df['FloorPos'] = df['Floor'] / df['HouseFloor']
```

Теперь посчитаем медиану, то есть, где в среднем расположены квартиры по отношению ко всему 

```python
floor_pos_norm = df.loc[df['Floor'] < df['HouseFloor'], 'FloorPos']
```

```python
plt.figure(figsize=(10, 4))
sns.kdeplot(floor_pos_norm)
plt.title('Распределение "этажности" квартиры относительно всего дома\n')
plt.xlabel(f'FloorPos\n median: {np.round(floor_pos_norm.median(), 3)}')
plt.show()
```

На графике видим 3 самых частых значения "этажности" квартиры. А медиана 0.556. думаю, в общем, можно заполнить Floor такими значениями, чтобы они были короче по середине дома $Floor = 0.556*HouseFloor$

Мне теперь стало интересно, как распределены разницы между значениями этажей, т.е. на сколько сильны такого рода выбросы.

```python
sns.kdeplot((df.loc[df['FloorPos'] > 1, 'Floor'] - df.loc[df['FloorPos'] > 1, 'HouseFloor']))
plt.title('Распределение превышения количества этажей в доме\n')
plt.xlabel(f"median: {(df.loc[df['FloorPos'] > 1, 'Floor'] - df.loc[df['FloorPos'] > 1, 'HouseFloor']).median()}")
plt.show()
```

Вот только меня смущает то, что медиана равна 2, то есть в большинстве (а если судить по графику, то сильно в большинстве) случаев выбросов такого рода (этаж квартиры больше, чем этажей всего в доме) этаж квартиры только на немного больше, чем этажей всего. Поэтому я просто поменяю местами такие значения

```python
must_be_housefloor = df.loc[df['FloorPos'] > 1, 'Floor']
df.loc[df['FloorPos'] > 1, 'Floor'] = df.loc[df['FloorPos'] > 1, 'HouseFloor']
df.loc[df['FloorPos'] > 1, 'HouseFloor'] = must_be_housefloor
```

```python
df.loc[df['FloorPos'] > 1, ['Floor', 'HouseFloor']]
```

Поменяли местами. И теперь пересчитаем FloorPos для нормальных и адекватных Floor и HouseFloor

```python
df['FloorPos'] = df['Floor'] / df['HouseFloor']
```
